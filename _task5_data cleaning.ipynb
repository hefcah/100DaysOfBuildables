```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\ma007\OneDrive\Desktop\archive (5)\Housing.csv"

df = pd.read_csv(file_path)
```

```python

print("First 5 Rows:")
print(df.head(5))

# Data types
print("\nData Types:")
print(df.dtypes)
```

```python

df = df.dropna()  
for col in df.columns:
    if df[col].dtype == "object":
        try:
            df[col] = pd.to_numeric(df[col], errors="ignore")
        except:
            pass
```

```python
import math

# Mean
def mean(vals):
    return sum(vals) / len(vals)

# Median
def median(vals):
    sorted_list = sorted(vals)
    n = len(sorted_list)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_list[mid-1] + sorted_list[mid]) / 2
    else:
        return sorted_list[mid]

# Mode
def mode(vals):
    freq_dict = {}
    for v in vals:
        if v in freq_dict:
            freq_dict[v] += 1
        else:
            freq_dict[v] = 1
    max_freq = max(freq_dict.values())
    result = [k for k, v in freq_dict.items() if v == max_freq]
    if len(result) == len(freq_dict):
        return None
    return result if len(result) > 1 else result[0]

# Variance
def variance(vals):
    m = mean(vals)
    return sum((x - m) ** 2 for x in vals) / len(vals)

# Standard Deviation
def std_dev(vals):
    return math.sqrt(variance(vals))
```

```python
if 'price' in df.columns:
    prices = df['price'].tolist()
    print("\nStats for Price Column:")
    print("Mean:", mean(prices))
    print("Median:", median(prices))
    print("Mode:", mode(prices))
    print("Variance:", variance(prices))
    print("Standard Deviation:", std_dev(prices))
```

```python
if 'price' in df.columns:
    plt.hist(df['price'], bins=30, edgecolor="black")
    plt.title("Histogram of House Prices")
    plt.xlabel("Price")
    plt.ylabel("Frequency")
    plt.show()

if 'furnishingstatus' in df.columns:
    df['furnishingstatus'].value_counts().plot(
        kind="bar", color="skyblue", edgecolor="black"
    )
    plt.title("Furnishing Status Counts")
    plt.xlabel("Furnishing Status")
    plt.ylabel("Count")
    plt.show()
```
